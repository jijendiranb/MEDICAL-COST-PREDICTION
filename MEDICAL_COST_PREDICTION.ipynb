```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv('/content/insurance.csv')
df.head(10)
```

```python
df.dtypes
```

```python
tem=pd.get_dummies(df['region'])
tem.head(10)
```

```python
df.drop('region',axis=1,inplace=True)
```

```python
df.head(10)
```

```python
#combining both dataset
df=pd.concat([df,tem],axis=1)
df.head(10)
```

```python
#coverting catagorical variables into numerical values
map={'yes':1,'no':0}
df['smoker']=df['smoker'].map(map)
df.head(10)


```

```python
map1={'male':1,'female':0}
df['sex']=df['sex'].map(map1)
df.head(10)
```

```python
df.corr()
```

```python
#correlation heatmap
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidths=2)
plt.show()
```

```python
#scatter plot
x=df['smoker']
y=df["charges"]
plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.xlabel('Non smokervs smoker')
plt.ylabel('charges')
plt.show()
```

Its clearly visible that smokers have high medical cost (13000-60000) than non smokers(0-38000).

```python
y=df['charges']
x=df.drop(['charges'],axis=1)
```

```python
x
```

```python
y
```

```python
#splitting the data for training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.15)
```

```python
#Training the multiple linear regression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
# checking the R squarred error
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#storing x in temporary variable
p_x=x
```

```python
from sklearn.preprocessing import PolynomialFeatures
#changing the data to a 3rd degree polynomial
pol=PolynomialFeatures(degree=3)
p_x=pol.fit_transform(x)
p_x
```

```python
# training the data similarly but with 3rd degree polynomial
x_train,x_test,y_train,y_test=train_test_split(p_x,y,random_state=1,test_size=0.15)
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# cross validation
from sklearn.model_selection import cross_val_score
c= cross_val_score(reg,p_x,y,cv=4)
c
```

```python
#final mean accuracy
print('Mean accuracy after cross validation is',c.mean()*100,end="%")
```

#conclusion
we have succesfully identified the pattern in the data and the 3rd degree polynomial regression has given a good accuracy of 82% on cross validation.
